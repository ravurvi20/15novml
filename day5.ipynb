```python
def image_show(file_name):
    img = cv2.imread(file_name)
    cv2.imshow(f'file name is {file_name}',img)
    key = cv2.waitKey(0)
    if key == ord("q"):
        cv2.destroyAllWindows()
    else:
        cv2.destroyAllWindows()
```

```python
image_show("photo2.jpg")
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imshow("test_window",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

vid = cv2.VideoCapture(0)
flag2 = True

while flag2:
    flag,img = vid.read()
    if flag:
        cv2.imshow("webcam", img)
        key = cv2.waitKey(0)
        if key == ord("q"):
            cv2.destoryAllWindows()
    else:
        break

vid.release()
```
